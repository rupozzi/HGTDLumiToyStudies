```python
import ROOT
f = ROOT.TFile.Open("HGTD_plots.root", 'READ')
c = []
h = []
f.ls()
plots = ["Number of HGTD hits vs. pT Gen", "Number of HGTD hits vs. pT G4",
         "Number of HGTD hits 2 GeV Gen", "Number of HGTD hits 2 GeV G4",
         "Number of HGTD hits vs. eta 2D Gen", "Number of HGTD hits vs. eta 2D G4",
         "Number of HGTD hits Gen", "Number of HGTD hits G4",
         "Number of HGTD hits vs. pT 2D Gen", "Number of HGTD hits vs. pT 2D G4",
         "Number of HGTD hits vs. eta Gen" ]
for i in range (len(plots)):
    c.append(ROOT.TCanvas('canvas'+str(i), '', 500, 500))
    h.append(f.Get(plots[i]))
    h[i].Draw()
    c[i].Draw()
```
